```python
import numpy as np
import pandas as pd
import statistics as stats

ages = [21, 22, 24, 26, 28, 30, 32, 34, 32, 40]

print("Mean: ", np.mean(ages))
print("Medean: ", np.median(ages))
print("Mode: ",stats.mode(ages))

```

```python
data = {
    "Students": ["A", "B", "C", "D", "E"],
    "Height": [5.5, 5.8, 5.5, 6.0, 5.7]
}

df = pd.DataFrame(data)
print("Mean: ", df["Height"].mean())
print("Median: ", df["Height"].median())
print("Mode: ", df["Height"].mode())
```
