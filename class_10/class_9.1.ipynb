```python
import numpy as np
import matplotlib.pyplot as plt

hours_studied = np.array([1, 2, 3, 4, 5])
marks = np.array([60, 70, 80, 90, 95])
sleep_hours = np.array([10, 8, 6, 4, 2])

correlation1 = np.corrcoef(marks, sleep_hours)[1,0]
correlation2 = np.corrcoef(hours_studied, marks)[1, 0]
print("Correlation Coefficient1:", correlation1)
print("Correlation Coefficient2:", correlation2)

# Plot
plt.scatter(marks, sleep_hours)
plt.xlabel("Marks")
plt.ylabel("Sleeping Hours")
plt.title("Marks vs Sleeping Hours")
plt.show()

plt.scatter(hours_studied, marks)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Hours vs Marks")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n_tosses = 100
n_simulations = 1000  

tosses = np.random.binomial(n=n_tosses, p=0.5, size=n_simulations)

plt.hist(tosses, bins=range(n_tosses+2), density=True, alpha=0.6, color='PLUM', edgecolor='black', label="Simulated")

mu = n_tosses * 0.5  
sigma = np.sqrt(n_tosses * 0.5 * 0.5)
x = np.linspace(min(tosses), max(tosses), 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'BLUE', lw=2, label="Normal Approximation")

# Labels
plt.title("Distribution of Number of Heads in 100 Coin Tosses")
plt.xlabel("Number of Heads")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
