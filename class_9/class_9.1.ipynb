```python
# Task - 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

data = np.array([2, 4, 6, 10, 40, 200, 1000])

mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)
cv = (std_dev / mean) * 100
skewness = skew(data)
kurt = kurtosis(data)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Coefficient of Variation:", cv)
print("Skewness:", skewness)
print("Kurtosis:", kurt)

# Histogram
plt.hist(data, bins=3)
plt.title("Data display in Graph")
plt.show()
```

```python
# New dataset
data2 = np.random.randint(0, 200, 10)

mean2 = np.mean(data2)
std2 = np.std(data2)

# Z-score for max value
max_value = np.max(data2)
z_score = (max_value - mean2) / std2

print("Dataset:", data2)
print("Mean:", mean2)
print("Standard Deviation:", std2)
print("Z-score of max value:", z_score)

# Percentiles
print("25th Percentile:", np.percentile(data2, 25))
print("50th Percentile:", np.percentile(data2, 50))
print("75th Percentile:", np.percentile(data2, 75))

# Boxplot
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title("Boxplot")
plt.show()

# Histogram
plt.hist(data2, bins=10, edgecolor='yellow', color='black')
plt.title("Histogram")
plt.show()
```
