```python
import pandas as pd

# Series
s = pd.Series([10,20,30])
print("Series:\n", s)

# DataFrame
data = {'Name':['Ali','Sara','Ahmed'],
        'Age':[22,25,23],
        'Marks':[85,90,88]}

df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python
print("Column Selection:\n", df['Name'])
print("Row Selection:\n", df.loc[1])
print("Conditional:\n", df[df['Marks'] > 85])
```

```python
print("Head:\n", df.head())
print("Unique Names:", df['Name'].unique())
print("Value Counts:\n", df['Age'].value_counts())

print("Column Names:", df.columns)
print("Index:", df.index)

print("Sorted by Marks:\n", df.sort_values('Marks'))

print("Null Check:\n", df.isnull().sum())

df['Marks'] = df['Marks'].replace(85, 86)
print("After Replace:\n", df)

print("Drop Column:\n", df.drop('Age', axis=1))
```
